```python
# Q1 >> Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
from bokeh.plotting import figure, show, output_file
```

```python
x=np.linspace(0,10,100)
```

```python
y=np.sin(x)
```

```python
a=figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')
a.line(x, y, legend_label="sin(x)", line_width=2)
show(a)
```

```python
#Q3 >>>Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
```

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
output_notebook()
p_bar = figure(x_range=fruits, title="Bar Graph", x_axis_label='fruits', y_axis_label='counts')
p_bar.vbar(x=fruits, top=counts, width=0.9)
show(p_bar)
```

```python
#Q4. Create a Bokeh histogram to visualize the distribution of the given data.
```

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
p_hist = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')
p_hist.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')
show(p_hist)
```
